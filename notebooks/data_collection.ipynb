```python
import serial 
import pandas as pd 
import time 
from datetime import datetime
```

```python
PORT = "COM3" 
BAUD_RATE = 115200
```

```python
ser = serial.Serial(PORT, BAUD_RATE, timeout=1) 
print("Connected to Arduino") 
time.sleep(2)
```

```python
label = "noisy" 

# focus 
# noisy 
# dark
```

```python
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S") 
filename = f"{label}_{timestamp}.csv" 
print(filename)
```

```python
duration = 60

rows = []

start = time.time()

while time.time() - start < duration:

    line = ser.readline().decode("utf-8").strip()

    if not line:
        continue

    try:

        sound, light = line.split(",")

        rows.append([
            float(sound),
            float(light),
            label
        ])

    except:
        continue

print("Finished")
print("Rows collected:", len(rows))
```

```python
df = pd.DataFrame(
    rows,
    columns=[
        "sound",
        "light",
        "label"
    ]
)

df.to_csv(filename, index=False)

df.head()
```
